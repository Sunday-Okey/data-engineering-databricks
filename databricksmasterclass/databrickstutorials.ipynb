## Access Data

```python
app_id = dbutils.secrets.get(scope="sunnyscope", key="appsecret")
tenant_id = dbutils.secrets.get(scope="sunnyscope", key="tenantsecret")
secret = dbutils.secrets.get(scope="sunnyscope", key="mysecret")
```

```python
spark.conf.set("fs.azure.account.auth.type.datalakeokey.dfs.core.windows.net", "OAuth")

spark.conf.set("fs.azure.account.auth.type.datalakeokey.dfs.core.windows.net", "OAuth")

spark.conf.set(
    "fs.azure.account.oauth.provider.type.datalakeokey.dfs.core.windows.net",
    "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",
)

spark.conf.set(
    "fs.azure.account.oauth2.client.id.datalakeokey.dfs.core.windows.net",
    app_id,
)

spark.conf.set(
    "fs.azure.account.oauth2.client.secret.datalakeokey.dfs.core.windows.net",
    secret,
)

spark.conf.set(
    "fs.azure.account.oauth2.client.endpoint.datalakeokey.dfs.core.windows.net",
    f"https://login.microsoftonline.com/{tenant_id}/oauth2/token",
)
```
