## Transforming Data with Pyspark DataFrame API

```python
%sql
USE CATALOG dbacademy;

USE SCHEMA labuser14614688_1776181907;
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
# Read in the customers data
customers_df = spark.read.format('csv')\
  .option('header', 'true')\
  .option('inferSchema', 'true')\
  .load('/Volumes/dbacademy/labuser14614688_1776181907/bronze_volume/customers/dim_customers.csv')
```

```python
# Inpect the data
display(customers_df.limit(10))
```

```python
customers_df.printSchema()
```

```python
customers_df.withColumn("name", upper(col("name"))).display()
```

```python
# Fetch the domain from the email column
customers_df = customers_df.withColumn("domain", split(col("email"), "@")[1])
```

```python
customers_df.groupBy("domain").agg(count("*").alias("count"))\
                                   .orderBy(desc("count"))\
                                       .display()
```

```python
customers_df.groupBy("domain").agg(count("*").alias("count"))\
                                   .orderBy(desc("count"))\
                                       .display()
```

```python
customers_df = customers_df.withColumn('processed_date', current_timestamp())
```

```python
customers_df.write.format('delta').mode('overwrite').saveAsTable('customers_silver')
```

```python
df = spark.table('customers_silver')
df.display()
```

```python
# Read in the products data
products_df = spark.read.format('csv')\
  .option('header', 'true')\
  .option('inferSchema', 'true')\
  .load('/Volumes/dbacademy/labuser14614688_1776181907/bronze_volume/products/dim_products.csv')
```

```python
# Inpect the data
display(products_df)
```

```python
products_df = products_df.withColumn('processed_date', current_timestamp())
display(products_df)
```

```python
products_df.groupBy("category").agg(avg("price").alias("avg_price")).display()
```
