## DATA WAREHOUSING

## Staging Layer

```sql
CREATE OR REPLACE TABLE stg_sales 
AS 
SELECT * FROM orders 
```

### Transanformation

```sql
CREATE VIEW trans_sales
AS
SELECT * FROM stg_sales WHERE Quantity IS NOT NULL 
```

```sql
SELECT * 
FROM trans_sales;
```

### Core Layer 

##### DimCustomers

```sql
CREATE OR REPLACE TABLE DimCustomers 
(
  CustomerID INT,
  CustomerName STRING,
  CustomerEmail STRING,
  DimCustomersKey INT
)
```

```sql
CREATE OR REPLACE VIEW view_DimCustomers
AS 
SELECT T.*, row_number() over(ORDER BY T.CustomerID) as DimCustomersKey FROM 
(
SELECT 
  DISTINCT(CustomerID) as CustomerID,
  CustomerName,
  CustomerEmail
FROM 
  orderDWH.trans_sales
) AS T
```
